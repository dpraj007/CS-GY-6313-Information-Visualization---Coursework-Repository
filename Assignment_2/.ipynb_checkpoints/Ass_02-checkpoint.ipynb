```python
# Define the list of countries of interest and the years range to focus on
selected_countries = ['AFG', 'BGD', 'IND', 'BTN', 'MMR', 'LKA']
years = [str(year) for year in range(2012, 2022)]  # from 2012 to 2021

# Filter each dataset by country code and select relevant years
# Renewable Energy Data
renewable_energy_filtered = renewable_energy_data[renewable_energy_data['Country Code'].isin(selected_countries)]
renewable_energy_filtered = renewable_energy_filtered[['Country Name', 'Country Code'] + years]

# Primary Completion Rate Data
primary_completion_filtered = primary_completion_data[primary_completion_data['Country Code'].isin(selected_countries)]
primary_completion_filtered = primary_completion_filtered[['Country Name', 'Country Code'] + years]

# Poverty Headcount Data
poverty_filtered = poverty_data[poverty_data['Country Code'].isin(selected_countries)]
poverty_filtered = poverty_filtered[['Country Name', 'Country Code'] + years]

# Maternal Mortality Data
maternal_mortality_filtered = maternal_mortality_data[maternal_mortality_data['Country Code'].isin(selected_countries)]
maternal_mortality_filtered = maternal_mortality_filtered[['Country Name', 'Country Code'] + years]

# CO2 Emissions Data
co2_filtered = co2_data[co2_data['Country Code'].isin(selected_countries)]
co2_filtered = co2_filtered[['Country Name', 'Country Code'] + years]

# Display the first few rows of each filtered dataset to confirm
renewable_energy_filtered.head(), primary_completion_filtered.head(), poverty_filtered.head(), maternal_mortality_filtered.head(), co2_filtered.head()
```

```python
# Convert the data from wide format to long format for easier merging
# First, we reshape each DataFrame to have 'Country Code', 'Year', and 'Value' for each indicator

# Reshape renewable energy data
renewable_long = renewable_energy_filtered.melt(id_vars=['Country Name', 'Country Code'], 
                                                value_vars=years, 
                                                var_name='Year', 
                                                value_name='Renewable_Energy')

# Reshape primary completion data
primary_completion_long = primary_completion_filtered.melt(id_vars=['Country Name', 'Country Code'], 
                                                           value_vars=years, 
                                                           var_name='Year', 
                                                           value_name='Primary_Completion')

# Reshape poverty data
poverty_long = poverty_filtered.melt(id_vars=['Country Name', 'Country Code'], 
                                     value_vars=years, 
                                     var_name='Year', 
                                     value_name='Poverty_Headcount')

# Reshape maternal mortality data
maternal_mortality_long = maternal_mortality_filtered.melt(id_vars=['Country Name', 'Country Code'], 
                                                           value_vars=years, 
                                                           var_name='Year', 
                                                           value_name='Maternal_Mortality')

# Reshape CO2 emissions data
co2_long = co2_filtered.melt(id_vars=['Country Name', 'Country Code'], 
                             value_vars=years, 
                             var_name='Year', 
                             value_name='CO2_Emissions')

# Now merge all datasets on 'Country Code' and 'Year'
merged_data = renewable_long.merge(primary_completion_long, on=['Country Name', 'Country Code', 'Year'])
merged_data = merged_data.merge(poverty_long, on=['Country Name', 'Country Code', 'Year'])
merged_data = merged_data.merge(maternal_mortality_long, on=['Country Name', 'Country Code', 'Year'])
merged_data = merged_data.merge(co2_long, on=['Country Name', 'Country Code', 'Year'])

# Display the first few rows of the merged dataset
merged_data.head()
```

```python
# Fill missing values with the mean of each column
merged_data['Renewable_Energy'].fillna(merged_data['Renewable_Energy'].mean(), inplace=True)
merged_data['Primary_Completion'].fillna(merged_data['Primary_Completion'].mean(), inplace=True)
merged_data['Poverty_Headcount'].fillna(merged_data['Poverty_Headcount'].mean(), inplace=True)
merged_data['Maternal_Mortality'].fillna(merged_data['Maternal_Mortality'].mean(), inplace=True)
merged_data['CO2_Emissions'].fillna(merged_data['CO2_Emissions'].mean(), inplace=True)

# Check if there are any remaining missing values
missing_values_check = merged_data.isnull().sum()

# Display the DataFrame and missing values check
merged_data.head(), missing_values_check
```

```python
# Normalize each indicator column to a range between 0 and 1 for the parallel coordinates plot
normalized_data = merged_data.copy()
indicator_columns = ['Renewable_Energy', 'Primary_Completion', 'Poverty_Headcount', 'Maternal_Mortality', 'CO2_Emissions']

# Apply normalization
for column in indicator_columns:
    min_val = normalized_data[column].min()
    max_val = normalized_data[column].max()
    normalized_data[column] = (normalized_data[column] - min_val) / (max_val - min_val)

# Display the first few rows of the normalized dataset
normalized_data.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import parallel_coordinates
import numpy as np
# Set up the plot
plt.figure(figsize=(12, 8))
parallel_coordinates(normalized_data, 'Country Name', 
                     cols=indicator_columns, 
                     color=plt.cm.viridis(np.linspace(0, 1, len(normalized_data['Country Name'].unique()))),
                     alpha=0.5)

# Add plot details
plt.title('Parallel Coordinates Plot for Socio-Economic and Environmental Indicators (2012-2021)')
plt.xlabel('Indicators')
plt.ylabel('Normalized Values (0 to 1)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Convert the normalized DataFrame to JSON format suitable for D3.js
normalized_data_json = normalized_data.to_json(orient='records')

# Save this JSON to a file for D3.js usage
json_file_path = '/home/ubuntu/INFO_VIS/Assignment_2/normalized_data.json'
with open(json_file_path, 'w') as json_file:
    json_file.write(normalized_data_json)

json_file_path
```
